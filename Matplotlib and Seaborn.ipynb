# BAR GRAPH

```python
#to create bar graph
x = [1,2,3,4,5,6]
y = [3,5,6,8,9,3]
plt.bar(x,y)
```

# Histogram

```python
x1 = [1,2,4,5,7,8,12,14,16,3,4,5]
```

```python
#bins specifies no of peaks
plt.hist(x1,bins = 9)
```

```python
#using seaborn
sns.histplot(x1,bins = 9)
```

```python
#to specify labels title to histogram
plt.hist(x1)
plt.xlabel('This is x axis')
plt.ylabel('this is y axis')
plt.title('Histogram of x1')
plt.show()
```

# Distribution plot

```python
sns.distplot(x1)
```
