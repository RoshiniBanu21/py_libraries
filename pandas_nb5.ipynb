### Histogram

Histogram are Bar charts where y axis contains frequecy of a variable 

```python
# by default number of bins/bucket is set to 10
blood_sugar = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
plt.hist(blood_sugar, rwidth=0.8) 
```

```python
plt.hist(blood_sugar,bins=5, rwidth=0.8) 
```

```python
#to customize bins
plt.hist(blood_sugar,bins=[80,100,125,150,180,200], rwidth=0.8) 
```

```python
plt.hist(blood_sugar,bins=[80,100,125,150,180,200], rwidth=0.8, color='red') 
```

```python
plt.hist(blood_sugar,bins=[80,100,125,150,180,200], rwidth=0.8, color='magenta', histtype='step') 
```

```python
blood_sugar_men = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
blood_sugar_women = [67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100]
plt.xlabel("Sugar Level")
plt.ylabel("Number Of Patients")
plt.title("Blood Sugar Chart")

plt.hist([blood_sugar_men,blood_sugar_women], bins=[80,100,125,150], rwidth=0.95, color=['purple','orange'],label=['men','women'])
plt.legend()
```

```python
blood_sugar_men = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
blood_sugar_women = [67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100]
plt.ylabel("Sugar Level")
plt.xlabel("Number Of Patients")
plt.title("Blood Sugar Chart")

plt.hist([blood_sugar_men,blood_sugar_women], bins=[80,100,125,150], rwidth=0.95, color=['orange','cyan'],label=['men','women'],orientation='horizontal')
plt.legend(shadow=True)
plt.grid()
```

## Pie Charts

```python
exp_vals = [1400,600,300,410,250]
exp_labels = ["Home Rent","Food","Phone/Internet Bill","Car ","Other Utilities"]
```

```python
plt.pie(exp_vals,labels=exp_labels)
```

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels)
plt.show()
```

```python
#to increase the size of the pie chart use radius function
#to know the percentage use autopct
#to take pies out use explode
#to specify angle use start angle
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels,radius=1.5,autopct="1.1f%%",shadow = True,explode=[0,0.2,0,0.2,0.1],startangle=180)


plt.savefig("pie.pdf",bbox_inches='tight',pad_inches=2,transparent=True)
```
